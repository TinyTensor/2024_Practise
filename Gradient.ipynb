```python
#compute y_hat=x*a+b
def predict(x, a, b):
  return x*a+b
#find gradient
def gradient(y_hat, y, x):
  da = 2*x*(y_hat-y)
  db = 2*(y_hat-y)
  return (da,db)
#update weight
def update_weight(a, b, lr, da, db):
  a_new = a - lr*da
  b_new = b - lr*db
  return (a_new, b_new)

x=[5,24,48]; y=[18,29,41]; N=len(x)
a=0.04; b=-0.34;lr=0.001
n=100; losses = []
for epoch in range(n):
  (da_total, db_total) = (0,0)
  loss_total=0.0
  for i in range(N):
    #predict y_hat
    y_hat=predict(x[i],a,b)
    #compute loss
    loss=(y_hat-y[i])**2
    loss_total=loss_total+loss
    #find gradient for each samples(x,y)
    (da,db)=gradient(y_hat,y[i],x[i])
    #find sum gradient
    da_total=da_total+da
    db_total=db_total+db
  #for debug only
  losses.append(loss_total/N)
  #update (a,b)
  (a,b)=update_weight(a,b,lr,da_total/N,db_total/N)
print(f"y={a}x+{b}")
import matplotlib.pyplot as plt
plt.plot(losses)
plt.xlabel("iteration")
plt.ylabel("loss")
plt.show()
```

```python
import numpy as np
def loss(a,b,x,y):
    y_model=a*x+b
    error=y-y_model
    return (error**2).mean()
x=np.array([1,2,4]); y=np.array([18,29,41])
a=0; b=0; lr=0.001; n=10000
for epoch in range(n):
    y_model=a*x+b
    error=y-y_model
    print(f"a={a},b={b}, lan thu {epoch}, sai so {loss(a,b,x,y)}")
    a_gradient=-2*(x*error).mean()
    b_gradient=-2*error.mean()
    print(f"a_grad={a_gradient},b_grad={b_gradient}")
    a=a-lr*a_gradient
    b=b-lr*b_gradient

print(f"a={a},b={b},ham so y={a}x+{b}, sai so {loss(a,b,x,y)}")
```
