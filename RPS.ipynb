```python
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import pandas as pd
from PIL import Image
import numpy as np
import gdown
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from re import sub
from google.colab import drive
```

```python
#parameters
img_height, img_width = 30, 30
batch_size = 32
data_dir = '/content/drive/My Drive/DATA/DATA_AUGMENTATION'
CLASS_LABEL = ['Background', 'Paper', 'Rock', 'Scissors']
```

```python
train_datagen = ImageDataGenerator(
    rescale = 1./255, #scale value between 1 and 0 since pixel values in images are usually in the range [0, 255]
    #augmentation
    rotation_range = 20, #randomly rotates 20 degrees
    width_shift_range = 0.2, #randomly shift horizontally 20%
    height_shift_range = 0.2, #randomly shift vertically 20%
    zoom_range = 0.2, #randomly zoom out or in 20%
    horizontal_flip = True, #randomly flip the image
    fill_mode = 'nearest', #fill in case the imager is blank after roatate
)
```

```python
val_datagen = ImageDataGenerator(
    rescale = 1./255,
    validation_split = 0.2,
)
#applying augmentation only on the validation data
```

```python
#load trainning and validation data
train_generator = train_datagen.flow_from_directory(
#vut data vao day
    directory = data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset = 'training'
)
val_generator = val_datagen.flow_from_directory(
#vut data vao
    directory = data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset = 'validation'
)
```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(4, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)
```

```python
val_loss, val_accuracy = model.evaluate(val_generator)
print(f'Validation Loss: {val_loss}')
print(f'Validation Accuracy: {val_accuracy}')
```
